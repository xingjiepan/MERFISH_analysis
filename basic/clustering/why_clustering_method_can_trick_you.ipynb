This notebook shows that the leiden clustering algorithm always segment the data points into clusters no matter if the data actually have "cluster structures".

In this example, leiden is used to cluster evenly spaced grid points on a 2D square.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import scanpy as sc
sc.settings.n_jobs = 56
sc.settings.set_figure_params(dpi=180, dpi_save=300, frameon=False, figsize=(3, 2), fontsize=8, facecolor='white')
```

```python
%matplotlib inline
```

```python
# Generate an dataset as the grids on a square
grid_length = 200
xv, yv = np.meshgrid(range(grid_length), range(grid_length))
df = pd.DataFrame({'x': xv.reshape(-1), 'y': yv.reshape(-1)})
adata = sc.AnnData(df)
```

```python
%%time
sc.pp.neighbors(adata, n_neighbors=8, n_pcs=0)
sc.tl.umap(adata)
```

```python
%%time
resolution = 0.1

sc.tl.leiden(adata, resolution=resolution)
sc.pl.umap(adata, color='leiden')
sc.pl.scatter(adata, x='x', y='y', color='leiden', legend_loc='on data')
```
